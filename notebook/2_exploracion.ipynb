### An√°lisis de texto

```python
import nltk 
from nltk.corpus import stopwords
print(stopwords.words('spanish'))
```

```python
texto_compras = ''
for row in compras_df['CONCEPTO']:
    texto_compras = texto_compras + ' ' + row
```

```python
stopword_list = stopwords.words('spanish') + ['2020', 'cont', 'millar', 'dpp', 'est', 'op', 'ret', 'mty', 'icom']
```

```python
texto_compras = ' '.join([x for x in texto_compras.strip().lower().split(' ') if x not in stopword_list and len(x)>4])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color='white')
wc.generate(texto_compras)

plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
